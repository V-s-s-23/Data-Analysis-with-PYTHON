# 4. Find out what product are often sold together? 

```python
df.head()
```

```python
# in raw 2 and 3 order id is same and address are also same ,so they would be order together by the customer. 
# make a column--> product order together
```

```python
# create new data frame
new_df = df[df['Order ID'].duplicated(keep=False)]
new_df.head(15)
```

```python
new_df['group'] = new_df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
new_df.head(15)
```

```python
new_df = new_df[['Order ID','group']].drop_duplicates()
new_df.head(20)
```

```python
from collections import Counter
from  itertools import combinations
```

```python
count = Counter()
for rows in new_df['group']:
    x = rows.split(',')
    count.update(Counter(combinations(x,3)))
    
for keys,values in count.most_common(10):
    print(keys,values)
```
